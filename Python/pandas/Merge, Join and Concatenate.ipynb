# Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\hp\Desktop\Alex The Analyst\Python\Exercises\pandas\LOTR.csv")

df1
```

```python
df2 = pd.read_csv(r"C:\Users\hp\Desktop\Alex The Analyst\Python\Exercises\pandas\LOTR 2.csv")

df2
```

## Merge

```python
# Inner join gives the common value in both dataframe.
# Default merge is inner join (intersection), df1 = left dataframe and df2 = rightdata frame
df1.merge(df2, how = 'inner')
```

```python
# specifying the join type and basis 
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# All the values in both dataframe, will not duplicate the overlapping data
df1.merge(df2, how = 'outer')
```

```python
# All the values from left dataframe and overlapping data from right dataframe
df1.merge(df2, how = 'left')
```

```python
# All the values from right dataframe and overlapping data from left dataframe
df1.merge(df2, how = 'right')
```

```python
# Takes each value in left dataframe and compares it to each values in right dataframe
df1.merge(df2, how = 'cross')
```

## Join

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
# Setting an indexes as FellowshipID
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

## Concatenate

```python
# puts one dataframe on top of another
pd.concat([df1, df2])
```

```python
# only takes the columns that are same in both dataframe
pd.concat([df1, df2], join = 'inner')
```

```python
# takes all the columns in both dataframe
pd.concat([df1, df2], join = 'outer')
```

```python
# joining using axis
pd.concat([df1, df2], join = 'outer', axis = 1)
```
