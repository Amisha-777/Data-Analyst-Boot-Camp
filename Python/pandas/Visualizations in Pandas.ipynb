# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\Alex The Analyst\Python\Exercises\pandas\Ice Cream Ratings.csv")
df
```

```python
# Setting Date Column as index
df = df.set_index('Date')
df
```

```python
# gives all the style available for visualization
print(plt.style.available)

# Specifying the style that is to be used
plt.style.use('classic')
```

```python
# By default a line plot is created
df.plot()
```

## Line Plot

```python
df.plot(kind = 'line', subplots = True)
```

```python
# Giving Title, X-axis title and y-axis title
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Raings', ylabel = 'Scores')
```

## Bar Graph

```python
df.plot(kind = 'bar')
```

```python
# Specifying bar chart type
df.plot(kind = 'bar', stacked = True)
```

```python
# Specifying column 
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# Making a vertical bar chart
df.plot.barh(stacked = True)
```

## Scatter Plot

```python
# We need to defing x and y for scatter plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
# Customizing scatter plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'Green')
```

## Histograms

```python
df.plot.hist()
```

```python
df.plot.hist(bins = 10)
```

## Box Plot

```python
df.boxplot()
```

## Area Plot

```python
df.plot.area()
```

```python
df.plot.area(figsize = (10,5))
```

## Pie Chart

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```
