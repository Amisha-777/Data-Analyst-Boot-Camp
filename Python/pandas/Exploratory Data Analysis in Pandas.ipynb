# EDA in Pandas

```python
# Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the csv file
df = pd.read_csv(r"your_file_path")
df
```

```python
# Changing the display settings
pd.set_option('display.float_format', lambda x : '%.2f' % x)
df
```

```python
# Getting the data info
df.info()
```

```python
# Getting the high-level overview of the data
df.describe()
```

```python
# Finding the missing values in each columns
df.isnull().sum()
```

```python
# Viewing the unique value in each columns
df.nunique()
```

```python
# Viewing the Top 10 Countries by 2022 Population
df.sort_values(by = "2022 Population", ascending = False).head(10)
```

```python
# Looking at the correlation of the data
df.corr(numeric_only = True)
```

```python
# Visulizing the correlation into a heatmap
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams['figure.figsize'] = (5,7) # adjusting the fig size

plt.show()
```

## Looking for the continents that have grown faster in different ways

```python
# Grouping by the continent
df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
# Looking for the columns name in dataframe
df.columns
```

```python
# Grouping by the population data requird (Column) in continent
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
# Transposing the data for better view
df3 = df2.transpose()
df3
```

```python
# Creating a line graph
df3.plot(figsize = (13,5))
```

```python
# Creating a box plot (very useful for finding the outlier)
df.boxplot(figsize = (20,10))
```

```python
# Looking for specific data types
df.select_dtypes(include = 'object')
```
